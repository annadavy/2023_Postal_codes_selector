The CodesSelector tool was developed for the purpose of finding the user chosen number of postal codes spread over the territory of a country in a regular way. The initial application of this tool was webscaping but it can serve as base for different location analysies based on postal codes or geocoordinates (latitude and longitude). The tool is using k-means algorithm to group the postal codes according to their location and then to select the most central ones for each group. The particular model can be saved but if there are a little bit different codes or geocoordinates needed at every run this can be achieved by just running the script without saving the model.

```python
import src
from src import CodesSelector
import warnings
warnings.filterwarnings("ignore")
```

The CodeSelector class takes 2 parameters: postal codes file with a table of postal codes with geocoordinates and shapefiles for the particular country. The example here is based on data from Poland but it is possible to find files relative for any country on the internet.

```python
poland=CodesSelector('codes_poland_all.txt','Poland_AL20280')
```

The plot function takes two parameters (tuple with the size of the plot and the number of codes (code groupings) we would like to get). The plot helps to get an idea how the groupings are spread all over the country.

```python
poland.plot((10,10),20)
```

The get_codes function produces the table with the requested number of codes and their geocoordinates. It takes just the number of codes as parameter.

```python
poland.get_codes(20)
```
